```python
class Ciudad:
    def __init__(self, nombre="", poblacion=0, pais=""):
        self.nombre = nombre
        self.poblacion = poblacion
        self.pais = pais

    def leerdatos(self):
        print(f"\n--- Ingresando datos de la ciudad ---")
        self.nombre = input("Nombre de la ciudad: ")
        self.poblacion = input("Número de habitantes: ")
        self.pais = input("País al que pertenece: ")

    def mostrardatos(self):
        print("\n" + "🏙️  " * 3)
        print(f"INFORMACIÓN DE LA CIUDAD")
        print("-" * 25)
        print(f"Ciudad:    {self.nombre}")
        print(f"Población: {self.poblacion}")
        print(f"País:      {self.pais}")
        print("-" * 25)

# --- Inicio del simulador interactivo ---
print("REGISTRO GLOBAL DE CIUDADES")
print("---------------------------")

# Creación del objeto ciudad1 y captura de datos
ciudad1 = Ciudad()
ciudad1.leerdatos()

# Creación del objeto ciudad2 y captura de datos
ciudad2 = Ciudad()
ciudad2.leerdatos()

# Mostrar la información de ambos objetos
print("\n--- RESULTADOS DEL REGISTRO ---")
ciudad1.mostrardatos()
ciudad2.mostrardatos()

print("Software de Control de ciudad finalizado")
print("Desarrollado por: Ian Gutiérrez | NL: 22")
```
